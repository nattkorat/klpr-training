## convert to YOLOv5 Format

```python
# Convert COCO annotations to YOLO format
def coco_to_yolo(coco_bbox, image_width, image_height):
    x_min, y_min, width, height = coco_bbox
    center_x = (x_min + width / 2) / image_width
    center_y = (y_min + height / 2) / image_height
    normalized_width = width / image_width
    normalized_height = height / image_height
    return center_x, center_y, normalized_width, normalized_height
```

```python
img = data['images']
len(img)
```

```python
for i in range(len(ann)):
    for j in range(len(img)):
        if ann[i]['image_id'] == img[j]['id']:
            ann[i]['width'] = img[j]['width']
            ann[i]['height'] = img[j]['height']
            ann[i]['file_name'] = os.path.splitext(img[j]['file_name'])[0]
ann
```

```python
for i in range(len(ann)):
    ann[i]['yolo'] = coco_to_yolo(ann[i]['bbox'], ann[i]['width'], ann[i]['height'])
ann
```

```python
for d in ann:
    with open(f'plate_serial_labels/{d["file_name"]}.txt', 'a') as file:
        if d['category_id'] == 2:
            cat = 0
        else:
            cat = 1
        file.write(f"{cat} {d['yolo'][0]:.6f} {d['yolo'][1]:.6f} {d['yolo'][2]:.6f} {d['yolo'][3]:.6f}\n")
```

```python
db_img = '/home/nattkorat/Desktop/klpr_dataset/images'
img = os.listdir(db_img)

label = os.listdir('plate_serial_labels')
```

```python
# train test valid split function 80, 10, 10 by default
def train_test_valid_split(data: list, test_size = 0.1, valid_size = 0.1):
    random.shuffle(data)

    total_size = len(data)
    test_split = math.ceil(total_size * test_size)
    valid_split = math.ceil(total_size * valid_size)

    test_data = data[:test_split]
    valid_data = data[test_split:valid_split+test_split]
    train_data = data[valid_split+test_split:]

    return train_data, test_data, valid_data
```

```python
data = train_test_valid_split(label, test_size=0.15, valid_size= 0.15)

target = ['train', 'test', 'valid']

# copy data to train set
for i in range(len(data)):
    for dat in data[i]:
        file_name = os.path.splitext(dat)[0]
        shutil.copyfile(f'plate_serial_labels/{dat}',f'plate_ser_dataset/{target[i]}/labels/{dat}')
        shutil.copyfile(f'{db_img}/{file_name}.jpg', f'plate_ser_dataset/{target[i]}/images/{file_name}.jpg')
```

## Getting Data for Training Label on Plate (Place)

```python
place_ann = []
for place in data['annotations']:
    if place['category_id'] == 4:
        place_ann.append(place)

place_ann[:5]
```

```python
for i in range(len(place_ann)):
    for img in data['images']:
        if place_ann[i]['image_id'] == img['id']:
            place_ann[i]['file_name'] = img['file_name']
            place_ann[i]['width'] = img['width']
            place_ann[i]['height'] = img['height']
place_ann[:5]
```

```python
for i in range(len(place_ann)):
    place_ann[i]['yolo'] = coco_to_yolo(place_ann[i]['bbox'], place_ann[i]['width'], place_ann[i]['height'])
place_ann[:5]
```

```python
labels = []
for ann in place_ann:
    labels.append(ann['attributes']['label'])

labels = sorted(list(set(labels)))
labels
```

```python
for i in range(len(place_ann)):
    place_ann[i]['cat'] = labels.index(place_ann[i]['attributes']['label'])

place_ann[:1]
```

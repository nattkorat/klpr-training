```python
from skimage import io, filters, color
from skimage.filters import threshold_otsu
import numpy as np
```

```python
img = io.imread("yolov5/test.jpg")
io.imshow(img)
```

```python
gray = color.rgb2gray(img)
io.imshow(gray)
```

```python
thresh = gray.mean()
bin_img = gray > thresh
bin_img = bin_img.astype(int)

io.imshow(bin_img)
```

```python
from skimage.filters import try_all_threshold
fix, ax = try_all_threshold(gray, figsize=(10,8), verbose=False)
```
